```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Görüntüyü Oku ve Yapay Gürültü Ekle
# 0 parametresi gri seviyede okumayı sağlar
img = cv2.imread('cicek.jpg', 0)

# Tuz-Biber gürültüsü ekleme
noisy = img.copy()
prob = 0.05  # %5 gürültü oranı (farkı görmek için ideal)
noisy[np.random.random(img.shape) < prob] = 255
noisy[np.random.random(img.shape) < prob] = 0

# 2. 4 Farklı Filtrenin Karşılaştırılması
kutu    = cv2.blur(noisy, (5, 5))
gauss   = cv2.GaussianBlur(noisy, (5, 5), 0)
medyan  = cv2.medianBlur(noisy, 5)
bilater = cv2.bilateralFilter(noisy, 9, 75, 75)

# 3. Sobel ve Laplacian Kenar Algılama
# Gradyan şiddeti için Sobel X ve Y birleştirilir
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel_final = cv2.magnitude(sobelx, sobely)

laplacian = cv2.Laplacian(img, cv2.CV_64F)

# Sonuçları Görselleştirme (Matplotlib)
plt.figure(figsize=(20, 12))

# Filtre Kıyaslaması
titles = ['Gürültülü', 'Kutu', 'Gauss', 'Medyan', 'Bilateral']
images = [noisy, kutu, gauss, medyan, bilater]

for i in range(5):
    plt.subplot(2, 5, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i]); plt.axis('off')

# Kenar Algılama Kıyaslaması
plt.subplot(2, 5, 7)
plt.imshow(sobel_final, cmap='gray')
plt.title('Sobel Gradyan'); plt.axis('off')

plt.subplot(2, 5, 9)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian'); plt.axis('off')

plt.tight_layout()
plt.show()
```

## 4. Sonuç ve Değerlendirme

Yapılan deneyde elde edilen bulgular şunlardır:

* **Filtreleme Başarısı:** Tuz-Biber gürültüsü içeren görsellerde **Medyan Filtre**'nin, aritmetik ortalama alan Kutu ve Gauss filtrelerine göre çok daha üstün bir restorasyon başarısı sergilediği gözlemlenmiştir.
* **Kenar Algılama:** Sobel operatörü, nesne sınırlarını (özellikle dikey ve yatay hatları) Laplacian operatörüne göre daha belirgin ve gürültüden arındırılmış bir şekilde ortaya koymuştur.

**Sonuç:** Gürültü temizleme ve kenar belirginleştirme aşamaları, görüntünün karakteristiğine göre doğru filtre seçimi yapıldığında çok daha başarılı sonuçlar vermektedir.
